```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import excel file

df = pd.read_excel(r'C:\Users\kevin\OneDrive\Documents\Data Analyst Portofolio\Flight Delay\FlightDelayPythonProject.xlsx')
df
```

```python
df.info()
```

```python
# checking null values
df.isnull().sum()
```

```python
# replace null value to zero
for i in df :
    df[i]:pd.to_numeric(df[i], errors='coerce')
        
df = df.fillna(value = 0)
df.isnull().sum()
```

```python
df.info()
```

```python
# change the data type to int
df['arr_delay'] = df['arr_delay'].astype('int64')
df["carrier_delay"] = df["carrier_delay"].astype('int64')
df["weather_delay"] = df["weather_delay"].astype('int64')
df["nas_delay"] = df["nas_delay"].astype('int64')
df["security_delay"] = df["security_delay"].astype('int64')
df["late_aircraft_delay"] = df["late_aircraft_delay"].astype('int64')

df.info()
```

```python
# generate a new column called total delay to sum all of the delay
df['total_delay'] = df.iloc[:,8:].sum(axis=1)
df
```

```python
# using group by to see delay caused by each airlines
carrier_delay = df[["carrier_name", "total_delay"]].groupby('carrier_name', as_index=False).sum()
carrier_delay
```

```python
# checking total delay value
check = carrier_delay.iloc[:,1].sum(axis=0)
check
```

```python
# carrier with most delay
most_carrier_delay = carrier_delay.sort_values(by='total_delay', ascending = False).head()
most_carrier_delay
```

```python
# making a bar chart
plt.figure(figsize=(15,7))
x= most_carrier_delay['carrier_name']
height= most_carrier_delay['total_delay']
plt.bar(x,height)
plt.title('Airlines With The Most Delay',fontsize=15)
plt.xlabel('Airlines Name',fontsize=15)
plt.ylabel('Total Delays',fontsize=15)
plt.xticks(rotation = 270)

# labelling the chart
for x,y in zip(x,height):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,5), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

#plt.savefig('Airlines With The Most Delay.jpg')
plt.show()
```

```python
# using group by to see total delay per year
total_delay_per_year = df[["year", "total_delay"]].groupby('year', as_index=False).sum()
total_delay_per_year
```

```python
# making a line chart
plt.figure(figsize=(15,7))
x= total_delay_per_year['year']
y= total_delay_per_year['total_delay']
plt.title('Total Delay per Year',fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Total Delays',fontsize=15)
#plt.savefig('My Chart.jpg')
plt.plot(x,y, color = 'blue')

# labelling the chart
for x,y in zip(x,y):

    label = "{:.0f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,5), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    
plt.show()
```

```python
# using group by to see total delay caused by each category
total_delay_per_category = df[["arr_delay", "carrier_delay","weather_delay","nas_delay","security_delay","late_aircraft_delay"]]
total_delay_per_category
```

```python
# transpose the table
transpose_category = total_delay_per_category.transpose()
transpose_category
```

```python
# calculating the total delay caused by each category
transpose_category['total_delay'] = transpose_category.iloc[0:,0:].sum(axis=1)
transpose_category
```

```python
# sorting the total delay caused by each category
top5 = transpose_category[['total_delay']].sort_values(by='total_delay', ascending = False).head()
top5
```

```python
# making a bar chart
plt.figure(figsize=(15,7))
x= top5.index.values 
height= top5['total_delay']
plt.bar(x,height)
plt.title('Most Delay Cause per Category',fontsize=15)
plt.xlabel('Delay Category',fontsize=15)
plt.ylabel('Total Delays',fontsize=15)

# labelling the chart
for x,y in zip(x,height):

    label = "{:.0f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,5), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.show()
```
